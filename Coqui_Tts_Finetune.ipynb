# **Load Data**

```python
text_data = pd.read_csv("/content/drive/MyDrive/AI_Interviewer_Project/Labels/interview_transcripts_by_turkers.csv")
text_data.rename(columns={"p1": "file_name", 0: "text"}, inplace=True)
```

```python
audio_folder = "/content/drive/MyDrive/AI_Interviewer_Project/audio/Audio"
audio_files = [f for f in os.listdir(audio_folder) if f.endswith(".wav")]
audio_data = pd.DataFrame({
    "file_name": [os.path.splitext(f)[0].lower() for f in audio_files],
    "audio_path": [os.path.join(audio_folder, f) for f in audio_files]
})

text_data["file_name"] = text_data["file_name"].str.lower()
merged_data = pd.merge(text_data, audio_data, on="file_name", how="inner")
```

# **Clean text**

```python
from datasets import Dataset, Audio

# Convert to a Dataset object
dataset = Dataset.from_pandas(merged_data)

# Add an "audio" column by loading the .wav files from "audio_path"
dataset = dataset.cast_column("audio_path", Audio(sampling_rate=16000))

# Rename the "audio_path" column to "audio" if necessary
dataset = dataset.rename_column("audio_path", "audio")

# Resample audio to 16 kHz
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))

# Verify the dataset
print(dataset[0])  # Print the first example
print(dataset.column_names)  # Check column names
```

```python
# Get the exact name of the transcript column
transcript_column_name = dataset.column_names[1]

# Rename the transcript column to "text"
dataset = dataset.rename_column(transcript_column_name, "text")
```

```python
import re

def cleanup_text(text):
    # Remove Interviewer/Interviewee tags
    text = re.sub(r"Interviewer:|Interviewee:", "", text)
    text = text.replace("|", " ")

    # Normalize punctuation and special characters
    text = re.sub(r"\[.*?\]", " ", text)  # Replace [???] with space
    text = re.sub(r"\s+", " ", text)     # Replace multiple spaces with a single space
    text = re.sub(r"[^\w\s.,!?]", "", text)  # Remove unsupported characters

    # Replace accented characters
    replacements = [
        ("à", "a"), ("ç", "c"), ("è", "e"), ("ë", "e"),
        ("í", "i"), ("ï", "i"), ("ö", "o"), ("ü", "u"),
    ]
    for src, dst in replacements:
        text = text.replace(src, dst)

    # Convert to lowercase
    text = text.lower()
    return text

# Apply cleanup to the text column
dataset = dataset.map(lambda example: {"text": cleanup_text(example["text"])})
```

# **Save Data in Coqui TTS Format**

```python
# Create manifest file: Each line => wav_path|transcript
output_dir = "/content/drive/MyDrive/AI_Interviewer_Project/Coqui_data"
os.makedirs(output_dir, exist_ok=True)

with open(os.path.join(output_dir, "metadata.csv"), "w", encoding="utf-8") as f:
    for row in dataset:
        f.write(f"{row['audio']['path']}|{row['text']}\n")
```

```python
from TTS.utils.manage import ModelManager

model_name = "tts_models/en/ljspeech/tacotron2-DDC"
manager = ModelManager()
model_path = manager.download_model(model_name)
```

```python
config = Tacotron2Config()

# Update config for training on your data
config.audio['sample_rate'] = 16000
config.run_name = "fine_tune_tacotron2"
config.output_path = "/content/tts_out"
config.datasets = [BaseDatasetConfig(path="/content/drive/MyDrive/AI_Interviewer_Project/Coqui_data")]
config.num_loader_workers = 2
config.batch_size = 16
config.eval_batch_size = 16
config.num_epochs = 100
config.run_eval = True
config.test_delay_epochs = 5
config.save_step = 1000
config.eval_step = 500
config.print_step = 25
```

# **Fine-Tune the Model**

```python
from TTS.api import Trainer
trainer = Trainer(config)
trainer.fit()
```

```python
# Load training logs
log_path = "/content/tts_out/fine_tune_tacotron2/train_logs.txt"
losses = []
steps = []

with open(log_path) as f:
    for line in f:
        if "Step:" in line and "Train Loss" in line:
            step = int(line.split("Step:")[1].split()[0])
            loss = float(line.split("Train Loss:")[1].split()[0])
            steps.append(step)
            losses.append(loss)

# Plot training loss
plt.figure(figsize=(10, 5))
plt.plot(steps, losses, label="Training Loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Tacotron2 Fine-Tuning Loss Curve")
plt.legend()
plt.grid()
plt.show()
```
